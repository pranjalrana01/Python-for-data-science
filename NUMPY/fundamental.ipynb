ARRAY FUNCTIONS

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
# max/min/sum/prod

np.max(a1, axis=0)  # it gives max from every row
#np.sum(a1) gives the sum
#np.prod(a1) gives the product
```

```python
# mean/median/std/variance

np.mean(a1) #using axis we can find column wise or row wise mean
```

```python
# trignometric_funcs

np.sin(a1)
```

```python
# dot ptoduct

a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12, 24).reshape(4,3)
answer = np.dot(a2,a3)
answer
```

```python
# log and exponents

np.log(a1) #log value calculated
```
