BROADCASTING

it describes how numpy array treats arrays with different shapes during arithmetic operations


```python
a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a)
print(b)

print(a + b)
```

MATHEMATICAL FORMULAS 


```python
a = np.arange(10)
np.sin(a)
```

```python
# sigmoid 

def sigmoid(array):
    return 1/(1+np.exp(-(array)))

a = np.arange(10)
sigmoid(a)
```

```python
# mean squared error

actual = np.random.randint(1, 50, 25)
predicted = np.random.randint(1, 50, 25)

def mse(actual, predicted):
    return np.mean((actual - predicted)**2)

mse(actual,predicted)
```
