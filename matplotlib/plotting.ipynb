```python
df = pd.read_csv('datasets/batsman_season_record.csv')
df.head(5)
```

```python
x = np.arange(df.shape[0])
width = 0.25

plt.bar(x - width, df['2015'], width, label='2015')
plt.bar(x, df['2016'], width, label='2016')
plt.bar(x + width, df['2017'], width, label='2017')

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Year-wise Comparison')
plt.legend()
plt.show()
```

```python
plt.bar(df['batsman'], df['2017'], label='2017')
plt.bar(df['batsman'], df['2016'], bottom=df['2017'],label='2016')
plt.bar(df['batsman'], df['2015'], bottom = df['2016'] + df['2017'], label='2015')

plt.legend()
plt.show()
```
