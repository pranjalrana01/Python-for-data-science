```python
import pandas as pd
import numpy as np
```

```python
# series -- it is a 1-D array holding data of any type

country = ['India', 'Pakistan', 'nepal','Bangladesh']

pd.Series(country)
```

```python
# integer

runs = [13, 24, 45, 76, 34, 79]
pd.Series(runs)
```

```python
# custom index

marks = [67, 57, 87, 49]
subjects = ['Maths', 'Science', 'Hindi', 'English']

pd.Series(marks, index = subjects)
```

```python
# setting a name

marks = pd.Series(marks, index =subjects, name = "pranjal")
marks
```
